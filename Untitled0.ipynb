```python
##removing duplicates
df.drop_duplicates()
df=df = df.drop_duplicates()
df
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[^a-z0-9]+", "_", regex=True)  # replace any non-alphanumeric with "_"
    .str.strip("_")
)
print(df.columns.tolist())
```

```python
df = df[~df["state_ut"].str.contains("total", case=False, na=False)]
df = df[~df["state_ut"].str.contains("percentage", case=False, na=False)]
df = df[~df["state_ut"].str.contains("all india", case=False, na=False)]
```

```python
df["state_ut"] = df["state_ut"].str.strip()
```

```python
for col in df.columns:
    if col not in ["sl_no", "state_ut"]:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ""), errors="coerce")
```

```python
df = df.fillna(0)
```

```python
for col in df.columns:
    if col not in ["sl_no", "state_ut"]:
        df.loc[df[col] < 0, col] = 0
```

```python
df = df.reset_index(drop=True)
```

```python
plt.figure(figsize=(12,6))
state_cases = df.groupby("state")["cases_reported"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=state_cases.values, y=state_cases.index, palette="Reds_r")
plt.title("Top 10 States with Highest Reported Rape Cases")
plt.xlabel("Total Cases")
plt.ylabel("State")
plt.show()
```

```python
plt.figure(figsize=(10,6))
age_groups = [
    "child_victims_below_6",
    "child_victims_6_12",
    "child_victims_12_16",
    "child_victims_16_18",
    "adult_victims_18_30",
    "adult_victims_30_45",
    "adult_victims_45_60",
    "adult_victims_60_plus"
]

age_totals = df[age_groups].sum().sort_values(ascending=False)
sns.barplot(x=age_totals.values, y=age_totals.index, palette="viridis")
plt.title("Age-wise Distribution of Victims Across India")
plt.xlabel("Number of Victims")
plt.ylabel("Age Group")
plt.show()
```

```python
plt.figure(figsize=(6,6))
victim_totals = {
    "Child Victims": df["child_victims_total"].sum(),
    "Adult Victims": df["adult_victims_total"].sum()
}

plt.pie(victim_totals.values(), labels=victim_totals.keys(),
        autopct="%1.1f%%", startangle=90, colors=["#ff9999","#66b3ff"])
plt.title("Child vs Adult Victims Distribution")
plt.show()
```

```python
df["child_to_adult_ratio"] = df["child_victims_total"] / (df["adult_victims_total"] + 1)  # +1 to avoid division by zero
```

```python
df["severity_index"] = df["total_victims"] / (df["cases_reported"] + 1)
```

```python
df["pct_child_below_12"]=((df["child_victims_below_6"] + df["child_victims_6_12"]) / df["total_victims"])
```
